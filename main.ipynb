```python
import pandas as pd 
df1 = pd.read_json("file1.json")
df1
```

```python

# Remove rows where 'restaurants' column is NaN
df1 = df1.drop(index=df1[df1['restaurants'].isna()].index)

# Remove rows where 'restaurants' column is an empty string
df1 = df1.drop(index=df1[df1['restaurants'] == ''].index)

# Remove rows where 'restaurants' column is an empty list
df1 = df1[~df1['restaurants'].apply(lambda x: isinstance(x, list) and len(x) == 0)]

# Remove rows with empty 'Cuisine Image URL'
#df1 = df1[df1['Cuisine Image URL'].notna()]

# Drop the unnecessary columns (without inplace=True to avoid issues)
df1 = df1.drop(columns=["results_found", "results_shown", "results_start", "message", "code", "status"])
```

```python
df1
```

```python
import pandas as pd

# Initialize an empty list to store extracted data
restaurants_data = []

# Iterate over each row in the 'restaurants' column
for index, row in df1.iterrows():
    restaurant = row['restaurants']  # Access the restaurant list for this row
    
    # Iterate over each restaurant dictionary in the list
    for entry in restaurant:
        if isinstance(entry, dict):  # Ensure that entry is a dictionary
            # Access the 'restaurant' key and extract the relevant data
            res_data = entry.get('restaurant', {})
            
            # Extract the required fields and arrange in specified order
            restaurant_data = {
                'Restaurant ID': res_data.get('R', {}).get('res_id'),  # Restaurant ID
                'Restaurant name': res_data.get('name'),  # Restaurant Name
                'Location': res_data.get('location', {}).get('address'),
                'Cuisines': res_data.get('cuisines'),
                'Ratings': res_data.get('user_rating', {}).get('aggregate_rating'),
                'Average cost for two': res_data.get('average_cost_for_two'),
                'Price range': res_data.get('price_range'),
                'Online delivery': res_data.get('has_online_delivery'),
                'Table booking': res_data.get('has_table_booking'),
                'City': res_data.get('location', {}).get('city'),
                'Latitude': res_data.get('location', {}).get('latitude'),
                'Longitude': res_data.get('location', {}).get('longitude'),
                'Cuisine Image URL': res_data.get('featured_image')  # Cuisine Image URL
                
            }
            
            # Append the extracted restaurant data to the list
            restaurants_data.append(restaurant_data)

# Convert the list of extracted data into a DataFrame
f1 = pd.DataFrame(restaurants_data)

# Save the extracted DataFrame to a CSV file
csv_file_path = 'file1.csv'  # Replace with your desired file path
f1.to_csv(csv_file_path, index=False)

```

MODEL TRAINING FOR THE CLEANED DATAFRAME(MERGED_DF)

MULTI HOT ENCODING

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Initialize MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Step 1: Split the 'Cuisines' column by commas, so we get lists of cuisines
cuisines_expanded = merged_df['Cuisines'].str.split(', ')

# Step 2: Apply MultiLabelBinarizer to convert each list of cuisines to one-hot encoding
encoded_df = pd.DataFrame(mlb.fit_transform(cuisines_expanded), columns=mlb.classes_)

# Step 3: Concatenate the original dataframe with the one-hot encoded cuisines
final_df = pd.concat([merged_df, encoded_df], axis=1)

# Display the final DataFrame
final_df
```

```python
final_df.to_csv("Restaurant_data.csv")
```

K MEANS ALGORITHM

```python
final_df.set_index(['Restaurant name', 'Location', 'City','Cuisines','Cuisine Image URL'], inplace=True)
```

```python

# Now apply KMeans on the numeric columns
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(final_df)
```

```python
#wccs (within cluster sum of squares)
kmeans.inertia_
```

```python
wccs = []

for i in range(1,7):
  model = KMeans(n_clusters=i)
  model.fit(final_df)
  wccs.append(model.inertia_)

wccs
```

```python
#Elbow plotting
import seaborn as sns
import matplotlib.pyplot as plt

plt.plot(range(1,7),wccs)
plt.show()
```
